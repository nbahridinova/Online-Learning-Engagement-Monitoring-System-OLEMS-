# Online Learning Engagement Monitoring System (OLEMS)

This part is the main program for the program. It will load student engagement data, calculates engagement scores, categorizes students, and analyzes class performance

```python
#import matplotlib to create notebook visual
import matplotlib.pyplot as plt

#imports the analyzer class & processes students records
from analyzer import ClassAnalyzer
```

```python
#creates analyzer obj & loads students engagemnet CSV dataset
system = ClassAnalyzer()
system.load_csv("data/students.csv")
```

```python
#displays each student with their calc engagemnet score and their category level
print("All Students' Records:")
system.display_students()
```

```python
#calculates and dispplays the class average eng. score
print("Average Engagement Score:")
print(system.average_score())
```

```python
#finds a& displays students with highest eng score
print("Highest Engagement Student:")
print(system.highest_student())
```

```python
#counts the # of students are in each eng. category level
print("Engagement Category Levels:")
print(system.category_counts())
```

```python
#displays studnts that were categorized in low level
print("Low Engaged Students:")
for student in system.low_engagement_students():
    print(student)
```

```python
#creates a chart for visualization of students engagemnet scores
system.plot_scores()
```

```python
#creates a bar chart that shows how many students are in each category 
category_data = system.category_counts()
plt.bar(category_data.keys(), category_data.values())
plt.xlabel("Engagement Level Ctageory")
plt.ylabel("Number of Students")
plt.title("Number of Students in every Engagement Level Category")
```
